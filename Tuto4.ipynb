```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import x,y
```

```python
l = np.arange(0,301)
k = np.arange(0,301)
```

```python
P = 1.0*(l**0.75)*(k**0.25)
ax = plt.axes(projection='3d')
ax.plot3D(l,k,P,'blue')
plt.show()
```

# Question 2

```python
h = 4*(x**2) + y**2
h
```

### Domain for h = R
### Range for h = [0,+inf)

```python
x = np.arange(-300,301)
y = np.arange(-300,301)
```

#### Plotting Line Graph in 3D for h

```python
h = 4*(x**2) + y**2
ax = plt.axes(projection='3d')
ax.plot3D(x,y,h,'blue')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
plt.show()
```

#### plotting surface for h

```python
ax = plt.axes(projection='3d')
X, Y = np.meshgrid(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
h = 4*(X**2) + Y**2
ax.plot_surface(X,Y,h,color='blue')
plt.show()
```

#### plotting contour for h

```python
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, h,50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
