#### f(x) = tan(x)

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100000)
y = [math.tan(i) for i in x]
plt.ylim(-10, 10)
plt.plot(x, y)
plt.show()
```

#### f(x) = e^x

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100000)
y = [math.e ** i for i in x]
plt.ylim(-10, 10)
plt.plot(x, y)
plt.show()
```
